```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import ipympl
import random
%matplotlib notebook

plt.ion()
```

```python
a = []
with open("rays.txt") as f:
    lines = f.readlines()
    for line in lines:
        a.append([float(x) for x in line.split(',')[:-1]])
a
```

```python
soa = np.array(random.sample(a, 5))

X, Y, Z, U, V, W = zip(*soa)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([0, 1])
ax.set_ylim([1, 3.4])
ax.set_zlim([-3, -1])
plt.show()
```
